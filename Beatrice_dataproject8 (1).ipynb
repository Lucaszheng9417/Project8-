```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data from a CSV file
# There is a file with the timestamp and the link to the book, and there is a file with just the timestamp 
datatime = pd.read_csv('Beatrce_book_timestamp.csv')
dataSub = pd.read_csv('eatrice_bookdata.csv')
dataSub
```

```python
# This makes a new column time_between the 2 times and sorts it into oldest to newest.
datatime['timestamp'] = pd.to_datetime(datatime['timestamp'])

datatime.sort_values(by='timestamp', inplace=True)
# Calculate the time differences and create a new column 'time_between'
datatime['time_between'] = datatime['timestamp'].diff()
datatime
```

```python
## What is the longest time between timestamps? What is the average time between timestamps?

datatime['time_between'] = pd.to_timedelta(datatime['time_between'])

# Find the longest time between timestamps
longest_time = datatime['time_between'].max()

# Calculate average time between timestamps
average_time = datatime['time_between'].mean()

# Returns what it found
print(f"Longest time between timestamps: {longest_time}")
print(f"Average time between timestamps: {average_time}")
```

```python
# Extract day from the 'timestamp' column and makes a column for it and counts it and the amount of events that happened on that day
datatime['day'] = datatime['timestamp'].dt.day

# Find the day with the most timestamps
day_with_most_events = datatime['day'].value_counts().idxmax()
events_on_that_day = datatime['day'].value_counts().max()

print(f"Day with the most timestamps: {day_with_most_events}, Events on that day: {events_on_that_day}")
```

```python
# Extract month from the 'timestamp' column
datatime['month'] = datatime['timestamp'].dt.month

# Find the month with the most timestamps
month_with_most_events = datatime['month'].value_counts().idxmax()
events_in_that_month = datatime['month'].value_counts().max()

print(f"Month with the most timestamps: {month_with_most_events}, Events in that month: {events_in_that_month}")
```

```python
# This graph shows the relationship between the visits of the sight and the time so on the x label is time 24 hours and 
#you can see I most frequent time to visit was around lunch time at 1pm and a bit befor around 8-9 am with some outliers.

# The pd.to_datetime is converting the timestamps to standardized datetime format which it was in but it was errering out without this
datatime['timestamp'] = pd.to_datetime(datatime['timestamp'])
# Extract hour from the 'timestamp' column
datatime['hour'] = datatime['timestamp'].dt.hour

# Sort DataFrame by timestamp
datatime = datatime.sort_values(by='hour')

# Count occurrences for each hour
hourly_visit_counts = datatime['hour'].value_counts().sort_index()

# Plotting each visit throughout the day
plt.figure(figsize=(12, 6))
plt.scatter(hourly_visit_counts.index, hourly_visit_counts.values, marker='o', s=50)
plt.xlabel('Hour of the Day (24-hour clock)')
plt.ylabel('Visit Count')
plt.title('Visits Throughout the Day')
plt.xticks(range(24))  # Set x-axis ticks to represent 0-24 hours
plt.grid(True)
plt.show()
```
